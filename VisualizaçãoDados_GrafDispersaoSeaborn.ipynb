Visualização de Dados - Gráfico de Dispersão com seaborn

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
#carregando a base de dados
dados = pd.read_csv('co2.csv')
dados.head()
```

```python
#plotando gráfico de dispersão
srn.scatterplot(dados.conc, dados.uptake, hue = dados.Type)
```

```python
#selecionando registros específicos da base de dados
q = dados.loc[dados['Type'] == 'Quebec']
m = dados.loc[dados['Type'] == 'Mississippi']
```

```python
#plotando dois gráficos em uma mesma figura
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
#carregando outra base de dados
dados2 = pd.read_csv('esoph.csv')
dados2.head()
```

```python
#gráfico entre dois atributos do conjunto de dados
srn.catplot(x = 'alcgp', y = 'ncontrols', data = dados2, jitter = False)
```

```python
#plotando gráficos com agrupamento
srn.catplot(x = 'alcgp', y = 'ncontrols', data = dados2, col = 'tobgp')
```
