Visualização de Dados - Gráfico de Densidade

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carregando a base de dados
dados = pd.read_csv('trees.csv')
dados.shape
```

```python
#plotando histograma
plt.hist(dados.iloc[:,1], bins = 6)
```

```python
#plotando histograma com seaborn
sns.distplot(dados.iloc[:,1], hist = True, kde = False, bins = 6, color = 'blue', hist_kws = {'edgecolor':'black'})
```

```python
#plotando gráfico de densidade
#plotando histograma com seaborn
sns.distplot(dados.iloc[:,1], hist = False, kde = True, bins = 6, color = 'blue', hist_kws = {'edgecolor':'black'})
```

```python
#plotando histograma e gráfico de densidade juntos
#plotando histograma com seaborn
sns.distplot(dados.iloc[:,1], hist = True, kde = True, bins = 6, color = 'blue', hist_kws = {'edgecolor':'black'})
```
